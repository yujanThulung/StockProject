```python
import finnhub
import pandas as pd
from dotenv import load_dotenv
from datetime import datetime, timedelta
import time

load_dotenv()
api_key = os.getenv("FINNHUB_API_KEY")
finhub_client = finnhub.Client(api_key=api_key)

finhub_client = finnhub.Client(api_key="d267tl9r01qh25lm67vgd267tl9r01qh25lm6800")

companies = [
    # Tech
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'NVDA', 'NFLX', 'ADBE', 
    'ORCL', 'CSCO', 'IBM', 'INTC', 'AMD', 'CRM', 'TSLA',

     # Finance
    'JPM', 'BAC', 'WFC', 'GS', 'MS', 'C', 'BLK', 'SCHW',
    
    # Energy
    'XOM', 'CVX', 'BP', 'COP',
    
    # Consumer
    'KO', 'PEP', 'PG', 'WMT', 'COST', 'MCD', 'SBUX', 'NKE', 'TGT',
    
    # Automotive
    'F', 'GM', 'RIVN',
    
    # Healthcare
    'JNJ', 'PFE', 'MRK', 'ABBV', 'TMO', 'UNH', 'LLY',
    
    # Telecom
    'T', 'VZ', 'TMUS',
    
    # Fintech/E-commerce
    'PYPL', 'SQ', 'SHOP', 'UBER', 'LYFT', 'ABNB',
    
    # International
    'BABA', 'TSM', 'SONY'
]

all_news = []

to_date = datetime.now()
from_date = to_date - timedelta(days=90) # last 90 days 

for symbol in companies:
    print(f"Fetching news fro {symbol}...")

    try:
        news = finhub_client.company_news(
            symbol,
            _from=from_date.strftime('%Y-%m-%d'),
            to=to_date.strftime('%Y-%m-%d')
        )

        for article in news:
            if not article.get('headline'):
                continue
            all_news.append({
                'company':symbol,
                'headline':article.get('headline',''),
                'summary':article.get('summary',''),
                'source': article.get('source',''),
                'datetime': article.get('datetime',0),
                'category': article.get('category', ''),
                'related': article.get('related', ''),
                'url': article.get('url','')
            })
 
        time.sleep(1.2)
    except Exception as e:
        print(f"Error fetching {symbol}: {e}")
        continue

df_raw = pd.DataFrame(all_news)
print(f"\nTotal articles collected: {len(df_raw)}")
print(f"Articles per company:\n{df_raw['company'].value_counts()}")

# Save raw data
df_raw.to_csv('../data/raw_news_multiple_companies.csv', index=False)
```
